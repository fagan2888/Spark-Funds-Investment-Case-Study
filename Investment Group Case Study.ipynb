### *** Checkpoint 1: Data Cleaning 1 *** ###
Loading the companies and rounds data into two data frames companies and rounds2 respectively. 

```python
# Importing pandas and numpy packages for analysis

import numpy as np
import pandas as pd
```

```python
#Importing companies data into a dataframe from txt file

companies = pd.read_csv("D:\Course 1 - Introduction to Data Management\Group Project 1 - Investment Case Study\Downloads\\companies.txt",sep="\t", encoding = "ISO-8859-1")
companies.head()
```

```python
#Importing rounds2 data into a dataframe from csv file

rounds2 = pd.read_csv("D:\Course 1 - Introduction to Data Management\Group Project 1 - Investment Case Study\Downloads\\rounds2.csv",sep=",", encoding = "ISO-8859-1")
rounds2.head()
```

### *** Removing special characters from the columns in the dataframes *** ###

```python
# Removing special characters from the columns 'permalink' and 'name' in companies dataframe
# Using str.encode() and str.decode functions() to remove special characters 
companies[['permalink','name']] = companies[['permalink','name']].applymap(lambda x: str(x).encode('utf-8').decode('ascii', 'ignore'))
companies.head()
```

```python
# Removing special characters from the columns 'company_permalink' and 'funding_round_permalink' in rounds2 dataframe
# Using str.encode() and str.decode functions() to remove special characters 

rounds2[['company_permalink','funding_round_permalink']] = rounds2[['company_permalink','funding_round_permalink']].applymap(lambda x: str(x).encode('utf-8').decode('ascii', 'ignore'))
rounds2.head()
```

### *** How many unique companies are present in rounds2? *** ###

```python
# Finding no of unique companies present in rounds2 dataframe - in 2 ways
# 1. Converting panda series(lower case) into set datastructure to find unique values
# 2. Converting panda series(lower case) into list datastructure to find unique values with unique func
print(len(set(rounds2.company_permalink.str.lower())))
print(len(rounds2.company_permalink.str.lower().unique().tolist()))
```

### *** How many unique companies are present in companies? *** ###

```python
# Finding no of unique companies present in companies dataframe - in 2 ways
# 1. Converting panda series(lower case) into set datastructure to find unique values
# 2. Converting panda series(lower case) into list datastructure to find unique values with unique func
print(len(set(companies.permalink.str.lower())))
print(len(companies.permalink.str.lower().unique().tolist()))
```

### *** Merge the two data frames so that all variables (columns) in the companies frame are added to the rounds2 data frame. Name the merged frame master_frame. How many observations are present in master_frame? *** ###

```python
# Merging all columns of companies dataframe to the rounds2 dataframe joining by permalink=company_permalink 
# Before merging, converting contents of both merge columns to lower case.
rounds2['company_permalink']=rounds2['company_permalink'].str.lower()
companies['permalink']=companies['permalink'].str.lower()
master_frame = pd.merge(rounds2,companies, left_on = 'company_permalink', right_on = 'permalink', how='left')
master_frame
```

```python
# Checking the various characteristics of master_frame
print(master_frame.info())
print(master_frame.describe())
```

```python
# Checking the shape of dataframes after merge
master_frame.shape
```

```python
# Inspecting percentage of null values in each of the columns
pd.DataFrame(round(100*(master_frame.isnull().sum()/master_frame.shape[0])),dtype=int)
```

```python
# Null values for raised_amount_usd doesnt make any sense or help in analysis. Hence dropping all rows containing null for 'raised_amount_usd'
master_frame = master_frame[~np.isnan(master_frame['raised_amount_usd'])] 
master_frame.shape
```

```python
# There are null values in many of the unnecessary column that aren't required in this assignment for analysis. 
# Droping such columns

columns_to_drop=['homepage_url','state_code','region','city','founded_at','funding_round_code','permalink']
print(columns_to_drop)

master_frame = master_frame.drop(columns_to_drop,axis=1)
```

```python
# Inspecting percentage of null values in each of the columns
pd.DataFrame(round(100*(master_frame.isnull().sum()/master_frame.shape[0]),2))
```

```python
master_frame.shape
```

```python
# Null values for country_code and category_list doesnt make any sense or help in analysis. 
# Hence dropping all rows containing null for these columns
master_frame = master_frame[~pd.isnull(master_frame['country_code'])] 
master_frame = master_frame[~pd.isnull(master_frame['category_list'])]
master_frame.shape
```

```python
# Inspecting percentage of null values in each of the columns
pd.DataFrame(round(100*(master_frame.isnull().sum()/master_frame.shape[0]),2))
```

```python
# Percentage of original no of records(114949) retained after cleaning all null values
round(100*(master_frame.shape[0]/114949),2)
```

### *** Checkpoint 4: Sector Analysis 1 *** ###
Code for a merged data frame with each primary sector mapped to its main sector 

```python
# Creating a new column 'primary_sector' that holds the first string before the vertical bar from the column 'category_list'
master_frame['primary_sector'] = master_frame['category_list'].apply(lambda x:str(x).split("|")[0])
master_frame.head()
```

### *** Use the mapping file 'mapping.csv' to map each primary sector to one of the eight main sectors *** ###

```python
# Reading the mapping data into a dataframe
mapping = pd.read_csv(r"D:\Course 1 - Introduction to Data Management\Group Project 1 - Investment Case Study\Downloads\mapping.csv",sep=",", encoding = "ISO-8859-1")
mapping.head()
```

```python
# Importing 're' packages for analysis requiring regular exressions.
import re
# There are values in the category_list column of mapping database where the string 'na' is replaced by '0'. Eg 'A0lytics'
# However there are instances where '0' won't be replaced by 'na'. Eg 'Enterprise 2.0'
# Using re.compile to generate a pattern to replace the '0' character with 'na' where ever needed
reg = re.compile('([a-zA-Z]*)0([a-zA-Z]+)')
mapping['category_list'] = mapping['category_list'].str.replace(reg,r'\1na\2')
mapping
```

```python
# Converting contents of 'category_list' column of mapping dataframe to lower case for later merging
mapping['category_list'] = mapping['category_list'].str.lower()
```

```python
# Setting the category_list as index from the mapping dataframe
mapping.set_index('category_list',inplace=True)
```

```python
# Selecting the main sector based on values from the category list column
# The following code gives the column header based on a value(equal to 1) in each row. 

main_sector = pd.DataFrame(mapping.apply(lambda x: x.isin([1]), axis=1).apply(lambda x: pd.Series(list(mapping.columns[x]))[0], axis=1), columns = ["main_sector"])
main_sector
```

```python
# Joining the 2 dataframes to map the main sector for the primary sector of each investment
# Before joining, converting the 'primary_sector' column of master_frame dataframe to lower case.
master_frame['primary_sector'] = master_frame['primary_sector'].str.lower()
master_frame = pd.merge(master_frame,main_sector,left_on = 'primary_sector', right_on = 'category_list', how='inner')
master_frame
```

```python
# Inspecting percentage of null values in each of the columns of master_frame after merge
pd.DataFrame(round(100*(master_frame.isnull().sum()/master_frame.shape[0]),2))
```

### *** Checkpoint 5: Sector Analysis 2 *** ###
Find out the most heavily invested main sectors in each of the three countries (for funding type FT and investments range of 5-15 M USD).

####  Create three separate data frames D1, D2 and D3 for each of the three countries containing the observations of funding type FT falling within the 5-15 million USD range. The three data frames should contain:

All the columns of the master_frame along with the primary sector and the main sector

The total number (or count) of investments for each main sector in a separate column

The total amount invested in each main sector in a separate column

```python
# Identifying all investments with funding type = 'venture' in the range of 5-15 M USD for the country 'USA'
D1 = master_frame.loc[(master_frame['country_code']=='USA') & (master_frame['raised_amount_usd']>= 5000000) & (master_frame['raised_amount_usd']<= 15000000) & (master_frame['funding_round_type']=='venture')]

# Finding total number (or count) of investments for each main sector from the above dataframe
P1 = pd.DataFrame(D1.pivot_table(values = 'company_permalink', index = 'main_sector', aggfunc='count', margins='True', margins_name='Total')).rename(columns={'company_permalink':'ct_of_investments'})
D1 = pd.merge(D1,P1,on='main_sector',how='left')

# Finding total amount invested in each main sector in a separate column
Q1 = pd.DataFrame(D1.pivot_table(values = 'raised_amount_usd', index = 'main_sector', aggfunc='sum', margins='True', margins_name='Total')).rename(columns={'raised_amount_usd':'total_amt_invested'})
D1 = pd.merge(D1,Q1,on='main_sector',how='left')
D1
```

```python
# Identifying all investments with funding type = 'venture' in the range of 5-15 M USD for the country 'GBR'
D2 = master_frame.loc[(master_frame['country_code']=='GBR') & (master_frame['raised_amount_usd']>= 5000000) & (master_frame['raised_amount_usd']<= 15000000) & (master_frame['funding_round_type']=='venture')]

# Finding total number (or count) of investments for each main sector from the above dataframe
P2 = pd.DataFrame(D2.pivot_table(values = 'company_permalink', index = 'main_sector', aggfunc='count', margins='True', margins_name='Total')).rename(columns={'company_permalink':'ct_of_investments'})
D2 = pd.merge(D2,P2,on='main_sector',how='left')

# Finding total amount invested in each main sector in a separate column
Q2 = pd.DataFrame(D2.pivot_table(values = 'raised_amount_usd', index = 'main_sector', aggfunc='sum', margins='True', margins_name='Total')).rename(columns={'raised_amount_usd':'total_amt_invested'})
D2 = pd.merge(D2,Q2,on='main_sector',how='left')
D2
```

```python
# Identifying all investments with funding type = 'venture' in the range of 5-15 M USD for the country 'IND'
D3 = master_frame.loc[(master_frame['country_code']=='IND') & (master_frame['raised_amount_usd']>= 5000000) & (master_frame['raised_amount_usd']<= 15000000) & (master_frame['funding_round_type']=='venture')]

# Finding total number (or count) of investments for each main sector from the above dataframe
P3 = pd.DataFrame(D3.pivot_table(values = 'company_permalink', index = 'main_sector', aggfunc='count', margins='True', margins_name='Total')).rename(columns={'company_permalink':'ct_of_investments'})
D3 = pd.merge(D3,P3,on='main_sector',how='left')

# Finding total amount invested in each main sector in a separate column
Q3 = pd.DataFrame(D3.pivot_table(values = 'raised_amount_usd', index = 'main_sector', aggfunc='sum', margins='True', margins_name='Total')).rename(columns={'raised_amount_usd':'total_amt_invested'})
D3 = pd.merge(D3,Q3,on='main_sector',how='left')
D3
```
